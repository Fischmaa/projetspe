```python
import pandas as pd
import pandas_profiling
from datetime import datetime
pd.options.mode.chained_assignment = None #enable warning
```

```python
#read random train sample
train = pd.read_csv('../only_booking_train_sample.csv',index_col=0).dropna()
```

```python
train=train.reset_index()

train['date']=0

for i in range(len(train)) :
    train['date'][i]=datetime.strptime(train['date_time'][i],'%Y-%m-%d %H:%M:%S')

train['dayOfWeek']=0

for i in range(len(train)):
    train['dayOfWeek'][i]=train['date'][i].isoweekday()

train['year']=0

for i in range(len(train)):
    train['year'][i]=train['date'][i].year

train['month']=0

for i in range(len(train)):
    train['month'][i]=train['date'][i].month

train['day']=0

for i in range(len(train)):
    train['day'][i]=train['date'][i].day
```

```python
pandas_profiling.ProfileReport(train)
```
